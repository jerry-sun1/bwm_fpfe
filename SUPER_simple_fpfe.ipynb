```python
from scipy.special import gamma
from scipy.optimize import curve_fit
import scipy.stats as stats
def chisq(x, k = 2):
    prefactor = 1/(2**(k/2) * gamma(k/2))
    postfactor = x**(k/2 - 1) * np.exp(-x/2)
    
    return prefactor*postfactor

def nc_chisq_wrapper(x, dof, nc):
    return stats.ncx2.pdf(x, dof, nc, loc=0, scale=1)
```

```python
def bwm_delay(toas, pos, log10_h=-14.0, cos_gwtheta=0.0, gwphi=0.0, gwpol=0.0, t0=55000,
              antenna_pattern_fn=None):
    """
    Function that calculates the earth-term gravitational-wave
    burst-with-memory signal, as described in:
    Seto et al, van haasteren and Levin, phsirkov et al, Cordes and Jenet.
    This version uses the F+/Fx polarization modes, as verified with the
    Continuous Wave and Anisotropy papers.

    :param toas: Time-of-arrival measurements [s]
    :param pos: Unit vector from Earth to pulsar
    :param log10_h: log10 of GW strain
    :param cos_gwtheta: Cosine of GW polar angle
    :param gwphi: GW azimuthal polar angle [rad]
    :param gwpol: GW polarization angle
    :param t0: Burst central time [s]
    :param antenna_pattern_fn:
        User defined function that takes `pos`, `gwtheta`, `gwphi` as
        arguments and returns (fplus, fcross)

    :return: the waveform as induced timing residuals (seconds)
    
    taken from enterprise_extensions
    """

    # convert
    h = 10 ** log10_h
    gwtheta = np.arccos(cos_gwtheta)
    t0 *= 1

    # antenna patterns
    if antenna_pattern_fn is None:
        apc = utils.create_gw_antenna_pattern(pos, gwtheta, gwphi)
    else:
        apc = antenna_pattern_fn(pos, gwtheta, gwphi)

    # grab fplus, fcross
    fp, fc = apc[0], apc[1]

    # combined polarization
    pol = np.cos(2 * gwpol) * fp + np.sin(2 * gwpol) * fc

    # Return the time-series for the pulsar
    return pol * h * np.heaviside(toas - t0, 0.5) * (toas - t0)
```

```python
npsrs = 45

#initialize all global params
rot_freqs = []
spindowns = []
bwm_signals =[]
quadratics = []
bwm_templates = []
base_toases = []
psrlocs = []



# put bwm exactly halfway through 10 years
#log10_h=-14.0, cos_gwtheta=0.0, gwphi=0.0, gwpol=0.0
bwm_t0 = 10 * 365.25 * 24 * 3600 * (1/2)
bwm_costheta = np.pi/3
bwm_phi = 2*np.pi/24
bwm_pol = np.pi/2
bwm_strain = -15
        
for psr in range(npsrs):
    rot_freq = 50 + 200*np.random.random()
    rot_freqs.append(rot_freq)
    rot_pd = 1/rot_freq
    
    #simulate one obsvn per yr
    rotns_per_month = np.ceil(30*24*3600/rot_pd) 
    base_toas = np.arange(0, 120, 1) * rotns_per_month * rot_pd
    base_toases.append(base_toas)
    
    
    wn_series = np.random.normal(loc=0, scale=100e-9, size=(len(base_toas)))
    wn_serieses.append(wn_series)
    
    spindown = 10** np.random.uniform(-21,-19)
    spindowns.append(spindown)
    quadratic = [0.5*spindown*(toa - base_toas[0])**2 for toa in base_toas]
    quadratics.append(quadratic)
    
    bwm_template = []
    for toaidx, toa in enumerate(base_toas):
        if toa < bwm_t0:
            bwm_template.append(0)
        else:
            bwm_template.append(toa - bwm_t0)

    bwm_templates.append(bwm_template)

    psr_theta = np.random.uniform(low = 0, high=np.pi)
    psr_phi = np.random.uniform(low=0, high=2*np.pi)
    
    psrloc = hp.ang2vec(psr_theta, psr_phi)
    psr_bwm_delay = bwm_delay(base_toas, psrloc, log10_h = bwm_strain, cos_gwtheta=np.cos(bwm_costheta),
                                gwphi=bwm_phi, t0=bwm_t0)
    
    bwm_signals.append(psr_bwm_delay)
    
```
